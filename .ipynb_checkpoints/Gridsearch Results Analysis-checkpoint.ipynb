```python
subject_no = 1
gridsearch_results_json = f'grid search results/wavelet smoothing/subject_{subject_no}/results.json'

results_dict = post_analysis.turn_gridsearch_results_into_dict(gridsearch_results_json)
```

```python
optimal_parameters = post_analysis.get_optimal_parameters(results_dict)
```

```python
optimal_parameters
```

```python
source_reject_per_subject = [[0,1,3]]
chs_to_exclude = source_reject_per_subject[subject_no - 1]

n_ica_components = 10
ica_object,subject_recording = cleaning.fit_ica(subject_no,n_ica_components)

ica_object.exclude = chs_to_exclude
ica_object.apply(subject_recording);
```

```python
# By default running comparison 100 times
accs_with_aug,accs_without_aug = post_analysis.compare_with_no_augmentation(subject_no,subject_recording,optimal_parameters)
```

```python
diffs = np.array(accs_with_aug) - (accs_without_aug)
np.mean(diffs)
```

```python
plt.boxplot((accs_with_aug,accs_without_aug),labels = ('Augmented','Not Augmented'));

plt.ylabel('Evaluation Accuracy');

plt.title(f'Subject {subject_no}');
```

```python
#By default three ensemble members
ensemble_eval_accuracies = post_analysis.train_augmentation_ensemble(subject_no,subject_recording,optimal_parameters)
```

### Without Point showing no Augmented Data

```python

m,b = np.polyfit(percent_increase_lst[1:],accs_across_percentages[1:],1)
best_fit_line = m * percent_increase_lst[1:] + b

plt.plot(percent_increase_lst[1:], best_fit_line, '--',c='orange')

plt.ylabel('Evaluation Accuracy')
plt.xlabel('% Increase in dataset size')
plt.errorbar(percent_increase_lst[1:],accs_across_percentages[1:],yerr = standard_errors[1:],capsize = 4,fmt = 'o',c='blue');
```
