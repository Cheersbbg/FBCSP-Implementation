```python
from src import classifier,preprocessing,loaders,augmentation,cleaning

from sklearn.model_selection import KFold
from sklearn.svm import SVC

from tqdm import tqdm

import numpy as np

import matplotlib.pyplot as plt

import statsmodels.api as sm

from scipy.signal import welch
```

```python
def get_power_ratio(smoothed_trial):

    trial_freqs,trial_psd = welch(smoothed_trial,scaling = 'spectrum')

    trial_freqs *= fs

    f_mu_and_beta = trial_freqs[8:30]
    f_gamma = trial_freqs[30:60]

    psd_mu_and_beta = trial_psd[8:30]
    psd_gamma = trial_psd[30:60]

    power_gamma = np.trapz(psd_gamma,f_gamma)
    power_mu_and_beta = np.trapz(psd_mu_and_beta,f_mu_and_beta)

    cur_ratio = power_mu_and_beta / power_gamma

    return cur_ratio

def get_power_ratios_C3(smoothed_trials):

    #For dataset 2a
    trials_C3 = smoothed_trials[:,7]

    return np.mean([get_power_ratio(trial) for trial in trials_C3])

def get_power_ratios_C4(smoothed_trials):

    trials_C4 = smoothed_trials[:,11]
    return np.mean([get_power_ratio(trial) for trial in trials_C4])
```

```python
subject_no = 1

original_trials,original_labels = loaders.extract_motor_imagery_trials(subject_no)

n_ica_components = 10
ica_object,subject_recording = cleaning.fit_ica(subject_no,n_ica_components)

channels_to_exclude = [0,1,3]
ica_object.exclude = channels_to_exclude
ica_object.apply(subject_recording);

trials,labels = loaders.extract_motor_imagery_trials(recording = subject_recording)
```

```python
sigma_vals = np.linspace(start = 0.35,stop = 1,num = 21)
```

```python
percent_increase_dataset = 0.2

fs = 250
f_bands = [[f,f+3] for f in range(3,30,3)]
filter_bank_coefficients = preprocessing.compute_filter_bank_coefficients(fs,f_bands)

n_times = 100

corr_coefs_power = []
corr_coefs_sigma = []

for i in tqdm(range(n_times)):

    power_ratio_values = []

    test_accuracy_values = []

    for sigma in sigma_vals:
        
        kf = KFold(n_splits = 10)
    
        train_test_indices = kf.split(trials,labels)
    
        test_accuracies = []
    
        power_ratios = []
        
        for i,(train_indices,test_indices) in enumerate(train_test_indices):
    
            train_trials,test_trials = trials[train_indices],original_trials[test_indices]
            train_labels,test_labels = labels[train_indices],original_labels[test_indices]
    
            #train_trials = preprocessing.bandpass_filter_trials(train_trials,fs = 250,l_freq = 8,h_freq = 30)
            #test_trials = preprocessing.bandpass_filter_trials(test_trials,fs = 250,l_freq = 8,h_freq = 30)
    
            n_trials_to_augment = int(len(train_trials) * percent_increase_dataset)
    
            #Only augmenting after we split into train and test data -> Zero data leakage
            augmented_trials,augmented_labels = augmentation.augment_with_wavelet_smooth_D2(train_trials,
                                                                            train_labels,
                                                                            n_trials_to_augment,
                                                                            sigma)
    
            cur_fold_power_ratio = get_power_ratios_C3(augmented_trials[-n_trials_to_augment:])
            power_ratios.append(cur_fold_power_ratio)
            
    
            base_classifier = SVC(kernel = 'linear',probability = True, C = 0.01)
    
            classifiers,spatial_filters = classifier.train_classifier(augmented_trials,
                                                                                    augmented_labels,
                                                                                    base_classifier = base_classifier,
                                                                                    mode = 'CSP')
            test_predictions = classifier.make_predictions(test_trials,
                                                           classifiers,
                                                           spatial_filters,
                                                           mode = 'CSP')
    
    
            test_accuracy = np.sum(test_predictions == test_labels) / len(test_labels)
            test_accuracies.append(test_accuracy)
    
        mean_test_accuracy = np.mean(test_accuracies)
        mean_power_ratio = np.mean(power_ratios)

        power_ratio_values.append(mean_power_ratio)
        test_accuracy_values.append(mean_test_accuracy)
    
    corr_power_ratio_test_accuracy = np.corrcoef(power_ratio_values,test_accuracy_values)[0][1]
    corr_sigma_test_accuracy = np.corrcoef(sigma_vals,test_accuracy_values)[0][1]

    corr_coefs_power.append(corr_power_ratio_test_accuracy)
    corr_coefs_sigma.append(corr_sigma_test_accuracy)
```
